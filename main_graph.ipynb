```python
import torch
from torch_geometric.loader import DataLoader
from torch_geometric.nn import global_mean_pool
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import HeteroConv, SAGEConv

class HeteroGNN(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels, edge_types):
        super(HeteroGNN, self).__init__()
        self.edge_types = edge_types
        
        # Extract unique node types
        self.node_types = list(set([node for edge in edge_types for node in (edge[0], edge[2])]))
        
        self.linear1 = nn.ModuleDict({
            node_type: nn.Linear(in_channels[node_type], hidden_channels)
            for node_type in self.node_types
        })
        
        self.conv1 = self.create_hetero_conv(hidden_channels)
        
        self.linear2 = nn.ModuleDict({
            node_type: nn.Linear(hidden_channels, hidden_channels)
            for node_type in self.node_types
        })
        
        self.conv2 = self.create_hetero_conv(hidden_channels)
        
        self.final_linear = nn.Linear(hidden_channels * len(self.node_types), out_channels)

    def create_hetero_conv(self, hidden_channels):
        conv_dict = {}
        for edge_type in self.edge_types:
            conv_dict[edge_type] = SAGEConv((hidden_channels, hidden_channels), hidden_channels)
        return HeteroConv(conv_dict, aggr='sum')

    def forward(self, data):
        x_dict, edge_index_dict, batch_dict = data.x_dict, data.edge_index_dict, data.batch_dict
        
        x_dict = {node_type: self.linear1[node_type](x) for node_type, x in x_dict.items()}
        
        x_dict = self.conv1(x_dict, edge_index_dict)
        x_dict = {node_type: F.relu(self.linear2[node_type](x)) for node_type, x in x_dict.items()}
        x_dict = self.conv2(x_dict, edge_index_dict)

        pooled_x = []
        for node_type in self.node_types:
            pooled_x.append(global_mean_pool(x_dict[node_type], batch_dict[node_type]))
        
        global_repr = torch.cat(pooled_x, dim=1)
        out = self.final_linear(global_repr)
        
        return out
```

```python
def validate_and_adjust_indices(data):
    corrected_edge_index_dict = {}
    node_offsets = {node_type: 0 for node_type in data.node_types}
    
    for node_type in data.node_types:
        batch_size = data[node_type].batch.max().item() + 1 if data[node_type].batch.numel() > 0 else 0
        node_offsets[node_type] = batch_size
    
    for key, value in data.edge_index_dict.items():
        src_type, _, dst_type = key
        src_offset = node_offsets[src_type]
        dst_offset = node_offsets[dst_type]
        
        adjusted_edge_index = value.clone()
        adjusted_edge_index[0, :] %= data[src_type].x.size(0)  # Use modulo to wrap around
        adjusted_edge_index[1, :] %= data[dst_type].x.size(0)  # Use modulo to wrap around
        
        corrected_edge_index_dict[key] = adjusted_edge_index
    
    data.edge_index_dict = corrected_edge_index_dict
    return data
```

```python
from torch_geometric.transforms import ToUndirected, RemoveDuplicatedEdges, Compose

def undirected_edges(dataset):
    # Create a composition of transforms
    transform = Compose([
        ToUndirected(),  # First, make the graph undirected
        RemoveDuplicatedEdges()  # Then, remove any duplicated edges
    ])
    
    # Apply the transforms to each graph in the dataset
    return [transform(data.clone()) for data in dataset]

# Usage
train_dataset=undirected_edges(train_dataset)
validation_dataset=undirected_edges(validation_dataset)
test_dataset=undirected_edges(test_dataset)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import HeteroConv, SAGEConv, Linear, global_mean_pool

class hgnn_kpi2d(torch.nn.Module):
    def __init__(self, hidden_channels, out_channels, metadata, node_dims, edge_dims):
        super().__init__()
        
        self.node_types = metadata[0]
        self.edge_types = metadata[1]
        
        # Node embeddings
        self.node_embeddings = torch.nn.ModuleDict()
        for node_type, in_dim in node_dims.items():
            self.node_embeddings[node_type] = Linear(in_dim, hidden_channels)
            
        # Edge embeddings
        self.edge_embeddings = torch.nn.ModuleDict()
        for edge_type, edge_dim in edge_dims.items():
            self.edge_embeddings[edge_type] = Linear(edge_dim, hidden_channels)
        
        # Graph convolution
        conv_dict = {}
        for edge_type in metadata[1]:
            # Modified to use edge features
            conv_dict[edge_type] = SAGEConv(
                in_channels=(-1, -1),  # -1 indicates dynamic input size
                out_channels=hidden_channels,
                aggr='mean'
            )
        self.conv = HeteroConv(conv_dict, aggr='mean')
        
        # Output layer
        self.output = Linear(hidden_channels, out_channels)
        
    def forward(self, x_dict, edge_index_dict, edge_attr_dict, batch_dict):
        # Transform node features
        x_dict = {
            node_type: self.node_embeddings[node_type](x)
            for node_type, x in x_dict.items()
        }
        
        # Transform edge features
        edge_attr_transformed = {}
        for edge_type, edge_attr in edge_attr_dict.items():
            edge_attr_transformed[edge_type] = self.edge_embeddings[edge_type](edge_attr)
        
        # Graph convolution with edge features
        x_dict = self.conv(x_dict, edge_index_dict, edge_attr_transformed)
        x_dict = {key: F.relu(x) for key, x in x_dict.items()}
        
        # Global pooling for each node type
        x_pooled = []
        for node_type, x in x_dict.items():
            if node_type in batch_dict:
                batch = batch_dict[node_type]
                pooled = global_mean_pool(x, batch)
                x_pooled.append(pooled)
        
        # Combine all node types
        x = torch.mean(torch.stack(x_pooled), dim=0)
        
        # Final prediction
        out = self.output(x)
        
        return out
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import HeteroConv, SAGEConv, Linear, global_mean_pool
from torch.nn import Sequential, ReLU
class hgnn_kpi2d(torch.nn.Module):
    def __init__(self, hidden_channels, out_channels, metadata, node_dims, edge_dims):
        super().__init__()
        
        self.node_types = metadata[0]
        self.edge_types = metadata[1]
        self.hidden_channels = hidden_channels
        self.out_channels = out_channels
        
        # Node embeddings
        self.node_embeddings = torch.nn.ModuleDict()
        for node_type, in_dim in node_dims.items():
            self.node_embeddings[node_type] = Linear(in_dim, hidden_channels)
            
        # Edge embeddings
        self.edge_embeddings = torch.nn.ModuleDict()
        for edge_type, edge_dim in edge_dims.items():
            edge_key = f"{edge_type[0]}_{edge_type[1]}_{edge_type[2]}"
            self.edge_embeddings[edge_key] = Linear(edge_dim, hidden_channels)
            
        # Message MLPs
        self.message_mlps = torch.nn.ModuleDict()
        for edge_type in metadata[1]:
            edge_key = f"{edge_type[0]}_{edge_type[1]}_{edge_type[2]}"
            self.message_mlps[edge_key] = Sequential(
                Linear(hidden_channels * 2, hidden_channels),
                ReLU(),
                Linear(hidden_channels, hidden_channels)
            )
        
        # Graph convolution
        self.convs = torch.nn.ModuleDict()
        for edge_type in metadata[1]:
            edge_key = f"{edge_type[0]}_{edge_type[1]}_{edge_type[2]}"
            self.convs[edge_key] = SAGEConv(
                in_channels=hidden_channels,
                out_channels=hidden_channels,
                aggr='mean'
            )
        
        # Output layer
        self.output = Linear(hidden_channels, out_channels)
        
    def forward(self, x_dict, edge_index_dict, edge_attr_dict, batch_dict):
        # Ensure all input tensors are on the same device as the model
        device = next(self.parameters()).device
        
        # Move input dictionaries to device
        x_dict = {k: v.to(device) for k, v in x_dict.items()}
        edge_index_dict = {k: v.to(device) for k, v in edge_index_dict.items()}
        edge_attr_dict = {k: v.to(device) for k, v in edge_attr_dict.items()}
        batch_dict = {k: v.to(device) for k, v in batch_dict.items()}
        
        # Transform node features
        x_dict = {
            node_type: self.node_embeddings[node_type](x)
            for node_type, x in x_dict.items()
        }
        
        # Transform edge features
        edge_attr_transformed = {}
        for edge_type, edge_attr in edge_attr_dict.items():
            edge_key = f"{edge_type[0]}_{edge_type[1]}_{edge_type[2]}"
            
            # Get number of edges from edge_index
            num_edges = edge_index_dict[edge_type].size(1)
            
            # Reshape edge attributes if necessary
            if edge_attr.size(0) != num_edges:
                if edge_attr.size(0) == 1:
                    edge_attr = edge_attr.repeat(num_edges, 1)
                else:
                    if edge_attr.size(0) < num_edges:
                        edge_attr = edge_attr.repeat(
                            (num_edges + edge_attr.size(0) - 1) // edge_attr.size(0), 1
                        )[:num_edges]
                    else:
                        edge_attr = edge_attr[:num_edges]
            
            # Transform the edge features
            edge_attr_transformed[edge_type] = self.edge_embeddings[edge_key](edge_attr)
        
        # Process each edge type
        new_x_dict = {}
        for edge_type in self.edge_types:
            src_type, _, dst_type = edge_type
            edge_key = f"{edge_type[0]}_{edge_type[1]}_{edge_type[2]}"
            
            if edge_type in edge_index_dict:
                try:
                    # Get source features
                    src_features = x_dict[src_type]
                    edge_index = edge_index_dict[edge_type]
                    edge_features = edge_attr_transformed[edge_type]
                    
                    # Combine source node features with edge features
                    src_idx = edge_index[0]
                    src_node_features = src_features[src_idx]
                    combined_features = torch.cat([src_node_features, edge_features], dim=1)
                    
                    # Process combined features through MLP
                    message_features = self.message_mlps[edge_key](combined_features)
                    
                    # Update source features with processed features
                    src_features_updated = src_features.clone()
                    src_features_updated[src_idx] = message_features
                    
                    # Perform convolution with updated features
                    out = self.convs[edge_key](src_features_updated, edge_index)
                    
                    # Initialize or append to destination features
                    if dst_type not in new_x_dict:
                        new_x_dict[dst_type] = []
                    new_x_dict[dst_type].append(out)
                    
                except Exception as e:
                    print(f"Error processing edge type {edge_type}")
                    print(f"Source features shape: {src_features.shape}")
                    print(f"Edge index shape: {edge_index.shape}")
                    print(f"Edge features shape: {edge_features.shape}")
                    raise e
        
        # Combine results for each node type
        x_dict = {}
        for node_type, features_list in new_x_dict.items():
            if len(features_list) > 1:
                # If multiple messages, take the mean
                x_dict[node_type] = torch.mean(torch.stack(features_list), dim=0)
            else:
                # If single message, use it directly
                x_dict[node_type] = features_list[0]
                
        # Apply ReLU
        x_dict = {key: F.relu(x) for key, x in x_dict.items()}
        
        # Global pooling for each node type
        x_pooled = []
        for node_type, x in x_dict.items():
            if node_type in batch_dict:
                batch = batch_dict[node_type]
                pooled = global_mean_pool(x, batch)
                x_pooled.append(pooled)
        
        if not x_pooled:
            raise ValueError("No node types were found in batch_dict")
        
        # Combine all node types
        x = torch.mean(torch.stack(x_pooled), dim=0)
        
        # Final prediction
        out = self.output(x)
        
        return out
```

```python
from sklearn.preprocessing import StandardScaler
import networkx as nx
import numpy as np

def convert_and_scale_graph(nx_graph):
    data = HeteroData()
    
    # Ensure all nodes have a type
    for node, attrs in nx_graph.nodes(data=True):
        if 'type' not in attrs:
            nx_graph.nodes[node]['type'] = 'unknown'
    
    # Create node mappings
    node_mappings = {}
    for node, attrs in nx_graph.nodes(data=True):
        node_type = attrs['type']
        if node_type not in node_mappings:
            node_mappings[node_type] = {}
        if node not in node_mappings[node_type]:
            node_mappings[node_type][node] = len(node_mappings[node_type])
    
    # Convert nodes
    for node_type, mapping in node_mappings.items():
        nodes = list(mapping.keys())
        node_features = [nx_graph.nodes[n].get('features', []) for n in nodes]
        
        # Handle single-value features
        node_features = [f if isinstance(f, list) else [f] for f in node_features]
        
        if node_features and any(node_features):
            # Ensure all feature lists have the same length
            max_length = max(len(f) for f in node_features)
            node_features = [f + [0] * (max_length - len(f)) for f in node_features]
            
            # Convert to tensor and scale
            node_features_tensor = torch.tensor(node_features, dtype=torch.float)
            scaler = StandardScaler()
            scaled_features = scaler.fit_transform(node_features_tensor)
            data[node_type].x = torch.tensor(scaled_features, dtype=torch.float)
        else:
            # If no features, use one-hot encoding
            num_nodes = len(mapping)
            data[node_type].x = torch.eye(num_nodes)
        
        data[node_type].num_nodes = len(mapping)
    
    # Convert edges
    edge_index_dict = {}
    edge_attr_dict = {}
    
    for u, v, attrs in nx_graph.edges(data=True):
        edge_type = attrs.get('type', 'unknown')
        src_type = nx_graph.nodes[u]['type']
        dst_type = nx_graph.nodes[v]['type']
        
        key = (src_type, edge_type, dst_type)
        if key not in edge_index_dict:
            edge_index_dict[key] = [[], []]
            edge_attr_dict[key] = []
        
        edge_index_dict[key][0].append(node_mappings[src_type][u])
        edge_index_dict[key][1].append(node_mappings[dst_type][v])
        
        edge_features = attrs.get('features', [])
        # Handle single-value features for edges
        edge_features = edge_features if isinstance(edge_features, list) else [edge_features]
        edge_attr_dict[key].append(edge_features)
    
    # Process edge data
    for key in edge_index_dict.keys():
        edge_index = torch.tensor(edge_index_dict[key], dtype=torch.long)
        
        if edge_attr_dict[key] and any(edge_attr_dict[key]):
            # Ensure all edge feature lists have the same length
            max_length = max(len(f) for f in edge_attr_dict[key])
            edge_features = [f + [0] * (max_length - len(f)) for f in edge_attr_dict[key]]
            
            edge_features_tensor = torch.tensor(edge_features, dtype=torch.float)
            scaler = StandardScaler()
            scaled_edge_features = scaler.fit_transform(edge_features_tensor)
            edge_attr = torch.tensor(scaled_edge_features, dtype=torch.float)
        else:
            edge_attr = torch.ones((edge_index.shape[1], 1), dtype=torch.float)
        
        data[key].edge_index = edge_index
        data[key].edge_attr = edge_attr
    
    return data

# Usage
pyg_data = convert_and_scale_graph(combined_graph)
```
